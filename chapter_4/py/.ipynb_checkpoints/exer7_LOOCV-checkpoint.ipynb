### Fit a logistic regression model that predicts Direction using Lag1 and Lag2.

```python
Weekly = load_data("Weekly")
Weekly
```

```python
X = MS(["Lag1", "Lag2"]).fit_transform(Weekly)
y = Weekly["Direction"] == "Up"
results = sm.GLM(y, X, family=sm.families.Binomial()).fit()
summarize(results)
```

### (b) Fit a logistic regression model that predicts Direction using Lag1 and Lag2 using all but the first observation.

```python
model = MS(["Lag1", "Lag2"]).fit(Weekly.iloc[1:])
X = model.transform(Weekly.iloc[1:])
y = Weekly["Direction"].iloc[1:] == "Up"
results = sm.GLM(y, X, family=sm.families.Binomial()).fit()
summarize(results)
```

### (c) Use the model from (b) to predict the direction of the first observation. You can do this by predicting that the first observation will go up if P(Direction = "Up" | Lag1, Lag2) > 0.5. Was this observation correctly classified?

```python
X_valid = model.transform(
    Weekly[["Lag1", "Lag2"]].iloc[0].to_frame().T
    # pd.DataFrame(
    #     {"Lag1": [Weekly["Lag1"].iloc[0]], 
    #      "Lag2": [Weekly["Lag2"].iloc[0]]}
    # )
)
X_valid
prob = results.predict(X_valid)
pred = "Down"
if prob[0] > .5: 
    pred = "Up"
print("prediction: %s\ntruth: %s" %(
    pred, 
    str(Weekly["Direction"].iloc[0])
))
```
