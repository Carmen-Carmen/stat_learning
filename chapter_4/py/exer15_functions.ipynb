```python
def power3(x, a): 
    return x ** a
```

```python
import numpy as np
from matplotlib.pyplot import subplots
```

```python
x = np.array(range(1, 100))
y = power3(x, 3)

fig, ax = subplots(figsize=(8, 8))
ax.scatter(x, y)
ax.set_xlabel("x values")
ax.set_ylabel("y values: power3(x, 3)")
# ax.set_xscale("log")
# ax.set_yscale("log")
```

```python
def plot_power(x: np.array, a): 
    y = power3(x, a)
    fig, ax = subplots(figsize=(8, 8))
    ax.scatter(x, y)
    ax.set_yscale("log")
```

```python
x = np.array(range(1, 100))
plot_power(x, 5)
```
